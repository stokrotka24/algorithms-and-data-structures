```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_results')
x = shelf_file['x']
k_random = shelf_file['k_random']
k_greedy = shelf_file['k_greedy']
k_kruskal = shelf_file['k_kruskal']
k_prim = shelf_file['k_prim']

shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Number of steps")
plt.xlabel("n")
plt.ylabel("k")
plt.plot(x, k_random, color="red", label="random")
plt.plot(x, k_greedy, color="green", label="greedy")
plt.plot(x, k_kruskal, color="blue", label="kruskal")
plt.plot(x, k_prim, color="purple", label="prim")

plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
