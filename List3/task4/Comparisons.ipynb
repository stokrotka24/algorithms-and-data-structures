```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_results')
x = shelf_file['x']
c_quick = shelf_file['c_quick']
c_s_quick = shelf_file['c_s_quick']
c_dual = shelf_file['c_dual']
c_s_dual = shelf_file['c_s_dual']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons")
plt.xlabel("n")
plt.ylabel("c(n)")
plt.plot(x, c_quick, color="red", label="quicksort")
plt.plot(x, c_s_quick, color="purple", label="quicksort with select")
plt.plot(x, c_dual, color="green", label="dual pivot quicksort")
plt.plot(x, c_s_dual, color="blue", label="dual pivot quicksort with select")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
