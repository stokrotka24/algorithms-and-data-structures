```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('worst_asc_parsed_results')
x = shelf_file['x']
c_quick = shelf_file['c_quick']
c_s_quick = shelf_file['c_s_quick']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons for array [1, 2, ..., n]")
plt.xlabel("n")
plt.ylabel("c(n)")
plt.plot(x, c_quick, color="red", label="quicksort")
plt.plot(x, c_s_quick, color="purple", label="quicksort with select")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
