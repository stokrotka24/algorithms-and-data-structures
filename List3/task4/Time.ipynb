```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_results')
x = shelf_file['x']
t_quick = shelf_file['t_quick']
t_s_quick = shelf_file['t_s_quick']
t_dual = shelf_file['t_dual']
t_s_dual = shelf_file['t_s_dual']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Time")
plt.xlabel("n")
plt.ylabel("t(n) [seconds]")
plt.plot(x, t_quick, color="red", label="quicksort")
plt.plot(x, t_s_quick, color="purple", label="quicksort with select")
plt.plot(x, t_dual, color="green", label="dual pivot quicksort")
plt.plot(x, t_s_dual, color="blue", label="dual pivot quicksort with select")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
