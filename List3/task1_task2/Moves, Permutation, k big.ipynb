```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/move/permutation_randomized_select_k_big')
x = shelf_file['x']
avgs = shelf_file['avgs']
stddevs = shelf_file['stddevs']
points_x = shelf_file['points_x']
points_y = shelf_file['points_y']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Moves for randomized select")
plt.xlabel("n")
plt.ylabel("moves")
plt.plot(x, avgs, color="blue", label="average")
plt.plot(x, avgs-stddevs, color="orange", label="average +- stddev")
plt.plot(x, avgs+stddevs, color="orange")
plt.legend(loc="upper left")
plt.scatter(points_x, points_y, color="pink")
plt.grid(True)
plt.show()
```
