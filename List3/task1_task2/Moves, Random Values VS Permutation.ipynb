```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/move/random_values_randomized_select_k_small')
x = shelf_file['x']
r_avgs = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results/move/permutation_randomized_select_k_small')
p_avgs = shelf_file['avgs']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Moves for randomized select, k small")
plt.xlabel("n")
plt.ylabel("moves")
plt.plot(x, r_avgs, color="yellow", label="random values")
plt.plot(x, p_avgs, color="cyan", label="permutation")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
