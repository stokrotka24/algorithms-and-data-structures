```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/comp/permutation_randomized_select_k_small')
x = shelf_file['x']
s_avgs = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results/comp/permutation_randomized_select_k_medium')
x = shelf_file['x']
m_avgs = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results/comp/permutation_randomized_select_k_big')
x = shelf_file['x']
b_avgs = shelf_file['avgs']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Average number of comparisons for randomized select")
plt.xlabel("n")
plt.ylabel("comparisons")
plt.plot(x, s_avgs, color="green", label="k small")
plt.plot(x, m_avgs, color="blue", label="k medium")
plt.plot(x, b_avgs, color="red", label="k big")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
