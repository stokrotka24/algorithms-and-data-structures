```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/comp/v_small')
x = shelf_file['x']
comparisons = shelf_file['results']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons for binary search, v small")
plt.xlabel("n")
plt.ylabel("comparisons")
plt.plot(x, comparisons, color="blue")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/comp/v_small')
x = shelf_file['x']
s_comparisons = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/comp/v_medium')
m_comparisons = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/comp/v_big')
b_comparisons = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/comp/v_nonexistent')
n_comparisons = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/comp/v_random')
r_comparisons = shelf_file['results']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons for binary search")
plt.xlabel("n")
plt.ylabel("comparisons")
plt.plot(x, s_comparisons, color="green", label="v small")
plt.plot(x, m_comparisons, color="blue", label="v medium")
plt.plot(x, b_comparisons, color="red", label="v big")
plt.plot(x, n_comparisons, color="orange", label="v nonexistent")
plt.plot(x, r_comparisons, color="purple", label="v random")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
