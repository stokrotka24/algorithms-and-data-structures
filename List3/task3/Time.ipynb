```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/time/v_small')
x = shelf_file['x']
time = shelf_file['results']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Time for binary search, v small")
plt.xlabel("n")
plt.ylabel("time")
plt.plot(x, time, color="blue")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/time/v_small')
x = shelf_file['x']
s_time = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/time/v_medium')
m_time = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/time/v_big')
b_time = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/time/v_nonexistent')
n_time = shelf_file['results']
shelf_file.close()

shelf_file = shelve.open('parsed_results/time/v_random')
r_time = shelf_file['results']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Time for binary search")
plt.xlabel("n")
plt.ylabel("time")
plt.plot(x, s_time, color="green", label="v small")
plt.plot(x, m_time, color="blue", label="v medium")
plt.plot(x, b_time, color="red", label="v big")
plt.plot(x, n_time, color="orange", label="v nonexistent")
plt.plot(x, r_time, color="purple", label="v random")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
