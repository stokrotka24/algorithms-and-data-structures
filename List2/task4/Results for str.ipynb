```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k100_s_10_str')
x = shelf_file['x']
c_merge = shelf_file['c_merge']
c_hybrid = shelf_file['c_hybrid']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons for switch = 10, k = 100, str")
plt.xlabel("n")
plt.ylabel("c(n)")
plt.plot(x, c_merge, color="blue", label="merge")
plt.plot(x, c_hybrid, color="orange", label="hybrid")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
