```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_switch_results_k1000_n50')
x = shelf_file['x']
c_insert = shelf_file['c_insert']
c_merge = shelf_file['c_merge']
c_quick = shelf_file['c_quick']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons to n = 50, k = 1000, int")
plt.xlabel("n")
plt.ylabel("c(n)")
plt.plot(x, c_insert, color="green", label="insertion")
plt.plot(x, c_merge, color="blue", label="merge")
plt.plot(x, c_quick, color="red", label="quick")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
