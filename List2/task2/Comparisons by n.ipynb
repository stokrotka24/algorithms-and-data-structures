```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k1')
x = shelf_file['x']
c_insert = shelf_file['c_insert']
c_merge = shelf_file['c_merge']
c_quick = shelf_file['c_quick']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons by n, k = 1")
plt.xlabel("n")
plt.ylabel("c(n)/n")
plt.plot(x, c_insert/x, color="green", label="insertion")
plt.plot(x, c_merge/x, color="blue", label="merge")
plt.plot(x, c_quick/x, color="red", label="quick")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
