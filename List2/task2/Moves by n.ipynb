```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k1')
x = shelf_file['x']
s_insert = shelf_file['s_insert']
s_merge = shelf_file['s_merge']
s_quick = shelf_file['s_quick']
shelf_file.close()

plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(15, 10))
plt.title("Moves by n, k = 1")
plt.xlabel("n")
plt.ylabel("s(n)/n")
plt.plot(x, s_insert/x, color="green", label="insertion")
plt.plot(x, s_merge/x, color="blue", label="merge")
plt.plot(x, s_quick/x, color="red", label="quick")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
