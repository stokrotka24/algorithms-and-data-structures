```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k100')
x = shelf_file['x']
s_merge = shelf_file['s_merge']
s_quick = shelf_file['s_quick']
s_dual = shelf_file['s_dual']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Moves, k = 100")
plt.xlabel("n")
plt.ylabel("s(n)")
plt.plot(x, s_merge, color="blue", label="merge")
plt.plot(x, s_quick, color="red", label="quick")
plt.plot(x, s_dual, color="purple", label="dual")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
