```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k100')
x = shelf_file['x']
t_merge = shelf_file['t_merge']
t_quick = shelf_file['t_quick']
t_dual = shelf_file['t_dual']
shelf_file.close()


plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Time, k = 100")
plt.xlabel("n")
plt.ylabel("t(n)  [seconds]")
plt.plot(x, t_merge, color="blue", label="merge")
plt.plot(x, t_quick, color="red", label="quick")
plt.plot(x, t_dual, color="purple", label="dual")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
