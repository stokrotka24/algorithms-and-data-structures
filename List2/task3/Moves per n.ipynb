```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k100')
x = shelf_file['x']
s_merge = shelf_file['s_merge']
s_quick = shelf_file['s_quick']
s_dual = shelf_file['s_dual']
shelf_file.close()

plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(15, 10))
plt.title("Moves by n, k = 100")
plt.xlabel("n")
plt.ylabel("s(n)/n")
plt.plot(x, s_merge/x, color="blue", label="merge")
plt.plot(x, s_quick/x, color="red", label="quick")
plt.plot(x, s_dual/x, color="purple", label="dual")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import shelve
%matplotlib inline


shelf_file = shelve.open('parsed_k100')
x = shelf_file['x']
s_merge = shelf_file['s_merge']
s_quick = shelf_file['s_quick']
s_dual = shelf_file['s_dual']
shelf_file.close()

plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(15, 10))
plt.title("Moves by n * log(n), k = 100")
plt.xlabel("n")
plt.ylabel("s(n)/(n * log(n))")
plt.plot(x, s_merge/(x * np.log(x)), color="blue", label="merge")
plt.plot(x, s_quick/(x * np.log(x)), color="red", label="quick")
plt.plot(x, s_dual/(x * np.log(x)), color="purple", label="dual")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
