```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results/bst_insert')
x = shelf_file['x']
bst_time = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results/rbt_insert')
rbt_time = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results/splay_insert')
splay_time = shelf_file['avgs']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Average time of inserting element")
plt.xlabel("n")
plt.ylabel("time")
plt.plot(x, bst_time, color="green", label="BST")
plt.plot(x, rbt_time, color="red", label="RBT")
plt.plot(x, splay_time, color="blue", label="SplayTree")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
