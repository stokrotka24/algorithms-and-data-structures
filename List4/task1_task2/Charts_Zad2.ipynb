```python
import matplotlib.pyplot as plt
import numpy as np
import shelve

shelf_file = shelve.open('parsed_results_zad2/bst_first_elem_asc')
x = shelf_file['x']
bst_comp = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results_zad2/rbt_first_elem_asc')
rbt_comp = shelf_file['avgs']
shelf_file.close()

shelf_file = shelve.open('parsed_results_zad2/splay_first_elem_asc')
splay_comp = shelf_file['avgs']
shelf_file.close()

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 10))
plt.title("Comparisons in finding the smallest element after inserting elements in ascending order")
plt.xlabel("n")
plt.ylabel("comparisons")
plt.plot(x, bst_comp, color="green", label="BST")
plt.plot(x, rbt_comp, color="red", label="RBT")
plt.plot(x, splay_comp, color="blue", label="SplayTree")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
